# 학습 여부 확인

```python
def comparing_2_model_paramters(param1, param2):

    # 두 파라미터 비교
    params_key_list = list(param1.keys())
    stack_dict = dict()
    for i, param_key in enumerate(params_key_list):

        model1_one_layer_param = param1[param_key]
        model2_one_layer_param = param2[param_key]
        bool_array = (model1_one_layer_param == model2_one_layer_param).flatten()

        stack_dict[i] = np.unique(bool_array, return_counts=True)
        
    return stack_dict
```

```python
# parameter들을 가지고 온다.
before_params = load_pickle(f"{TEST_DATA_DIR}/before_weight2.pickle")
after_params = load_pickle(f"{TEST_DATA_DIR}/after_weight2.pickle")
```

```python
comparing_2_model_paramters(param1=before_params, param2=after_params)
```
